```python
x = np.linspace(0, 1, 1000)
# plt.figure(dpi=300, figsize=(10, 6))
fig, ax = plt.subplots(1, 1, dpi=300, figsize=(10, 6))
p = 0.8

mean, var, skew, kurt = norm.stats(moments='mvsk')
plt.figure(dpi=300, figsize=(10, 6))
ax.plot(x, p*norm(loc=0.25, scale=0.03).pdf(x)+(1-p)*norm(loc=0.5, scale=0.03).pdf(x),
       'r-', lw=5, alpha=0.6, label='norm pdf')
```

```python
fig, axs = plt.subplots(2, 2, dpi=300, figsize=(10, 6))
fig.suptitle('Multimodal Distribution with different volatility')
p=0.8
fx = p*norm(loc=0.25, scale=0.03).pdf(x)+(1-p)*norm(loc=0.5, scale=0.03).pdf(x)
axs[0, 0].plot(x, fx)
axs[0, 0].set_title('Mu=[0.25, 0.5], Sigma=[0.03, 0.03]')
fx = p*norm(loc=0.25, scale=0.1).pdf(x)+(1-p)*norm(loc=0.5, scale=0.1).pdf(x)
axs[0, 1].plot(x, fx)
axs[0, 1].set_title('Mu=[0.25, 0.5], Sigma=[0.1, 0.1]')
p = 0.2
fx = p*norm(loc=0.25, scale=0.03).pdf(x)+(1-p)*norm(loc=0.5, scale=0.03).pdf(x)
axs[1, 0].plot(x, fx)
axs[1, 0].set_title('Mu=[0.25, 0.5], Sigma=[0.03, 0.03]')
fx = p*norm(loc=0.25, scale=0.1).pdf(x)+(1-p)*norm(loc=0.5, scale=0.1).pdf(x)
axs[1, 1].plot(x, fx)
axs[1, 1].set_title('Mu=[0.25, 0.5], Sigma=[0.1, 0.1]')
fig.savefig('./fig/multimodal_dist.png')
```
